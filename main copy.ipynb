```python
import tkinter as tk
import cv2 as cv
import numpy as np
```

```python
def video(d, s, mode):
    cap = cv.VideoCapture(0)

    if not cap.isOpened():
        print("Cannot open camera")
        exit()

    delta = d

    if s>255:
        s=255

    if s<0:
        s=0

    frames_list = []
    
    while True:
        # Capture frame-by-frame
        #failsafes
        if delta < 1:
            delta = 1
        
        if s<240:
            s = 240

        if s > 255:
            s = 255

        ret, frame = cap.read()
    
        # if frame is read correctly ret is True
        if not ret:
            print("Can't receive frame (stream end?). Exiting ...")
            break
        # Our operations on the frame come here
        frames_list.append(frame)
        if len(frames_list) >= delta+2:
            frames_list = frames_list[1:]

        diff_image = cv.absdiff(frames_list[0], frames_list[-1])
        gray_image = cv.cvtColor(diff_image, cv.COLOR_BGR2GRAY)
        (thresh, mono) = cv.threshold(gray_image, 255 - s, 255, cv.THRESH_BINARY)
        
        # Display the resulting frame
        if mode == "mono":
            cv.imshow('capture', mono)

        elif mode == "red":
            pure = np.tile([0,0,1], (1080, 1920, 1)).astype(np.uint8)
            color_img = np.multiply(pure, np.expand_dims(mono, axis=-1))
            cv.imshow('capture', color_img)
        
        elif mode == "green":
            pure = np.tile([0,1,0], (1080, 1920, 1)).astype(np.uint8)
            color_img = np.multiply(pure, np.expand_dims(mono, axis=-1))
            cv.imshow('capture', color_img)

        elif mode == "blue":
            pure = np.tile([1,0,0], (1080, 1920, 1)).astype(np.uint8)
            color_img = np.multiply(pure, np.expand_dims(mono, axis=-1))
            cv.imshow('capture', color_img)

        elif mode == "noise":
            noise_img = np.random.randint(0, 2, (1080, 1920, 3), dtype=np.uint8)
            color_img = np.multiply(noise_img, np.expand_dims(mono, axis=-1))
            cv.imshow('capture', color_img)

        if cv.waitKey(1) == ord('q'):
            break
    
    # When everything done, release the capture
    cap.release()
    cv.destroyWindow('capture')
    cv.waitKey(1)
    return None

video(1, 252, "noise")
```
